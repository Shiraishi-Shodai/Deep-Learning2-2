### 計算② マニュアルでの計算

```python
def softmax(x):
    if x.ndim == 2:
        x_max = x.max(axis=1).values.reshape(-1, 1)
        x = x - x_max
        x_exp = torch.exp(x)
        z = x_exp / x_exp.sum(axis=1).reshape(-1, 1)
    
    elif x.ndim == 1:
        x_max = x.max().values
        x = x - x_max
        x_exp = torch.exp(x)
        z = x_exp / x_exp.sum()  
    return z
```

```python
z = softmax(pred)
z
```

```python
y = -torch.log(z[torch.arange(len(z)), label])
y.mean()
```
